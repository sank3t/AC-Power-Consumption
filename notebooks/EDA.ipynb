```python
import pandas as pd

import plotly.graph_objects as go

# Customizing pandas output
pd.set_option("display.max_rows", 100, "display.max_columns", 25)
```

```python
resampled_csv_to_df_lambda = lambda file_path: pd.read_csv(
    file_path,
    parse_dates=["timestamp"],
    index_col=["timestamp"]
)

ac_pwr_hourly_df = resampled_csv_to_df_lambda(file_path="../data/resampled/ac_pwr_hourly.csv")
ac_pwr_daily_df = resampled_csv_to_df_lambda(file_path="../data/resampled/ac_pwr_daily.csv")
ac_pwr_monthly_df = resampled_csv_to_df_lambda(file_path="../data/resampled/ac_pwr_monthly.csv")

ac_pwr_total_consumed_df = pd.read_csv("../data/resampled/ac_pwr_total_consumed.csv")
```
