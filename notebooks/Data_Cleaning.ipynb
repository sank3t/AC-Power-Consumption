```python
import os

import pandas as pd
```

```python
ac_pwr_df = pd.read_csv("../data/raw/AC_Data.csv")

display(ac_pwr_df.head())
```

## Renaming columns

```python
ac_pwr_df = ac_pwr_df.rename(columns={"0": "timestamp"})

ac_pwr_columns = ["_".join(col.lower().split(" ")) for col in ac_pwr_df.columns[1:]]
ac_pwr_df.columns = ["timestamp"] + ac_pwr_columns
```

```python
display(ac_pwr_df.head())
```

## Dumping the cleaned data

```python
os.makedirs("../data/cleaned/", exist_ok=True)
```

```python
ac_pwr_df.to_feather("../data/cleaned/ac_pwr_cleaned.feather")
```
